```python
import pandas as pd
```

```python
data=pd.read_csv('/content/Iris.csv')
```

```python
data.head()
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import scale
```

```python
features=scale(data[['SepalLengthCm',	'SepalWidthCm' ,	'PetalLengthCm',	'PetalWidthCm']])
```

```python
pd.DataFrame(features).describe()
```

```python
target=data['Species']
```

```python
target.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target=pd.Series(LabelEncoder().fit_transform(target))
```

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(features,target,test_size=0.15)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
bc=GaussianNB()
```

```python
bc.fit(x_train,y_train)
```

```python
pred=bc.predict(x_test)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,classification_report,accuracy_score
```

```python
print('accuracy:',accuracy_score(pred,y_test))
print('precision:',precision_score(pred,y_test,average='micro'))
print('recall:',recall_score(pred,y_test,average='micro'))
print('f1_score:',f1_score(pred,y_test,average='micro'))
```

```python
print(classification_report(pred,y_test))
```
